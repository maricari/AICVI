### Resolución

Para obtener las múltiples detecciones:
- Tomé como base las imágenes seleccionadas en el paso anterior (ya están en una escala adecuada).
- Para realizar la detección múltiple implementé la función `matchTemplateMulti`. Para cada imagen invoqué a la función utilizando el método que mejor funcionó en la detección simple.
- Para filtrar el solapamiento entre detecciones muy cercanas usé un algoritmo de NMS
- Guardé la imagen resultante en la carpeta `matched_multi`.


```python
import cv2 as cv
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (6, 6)

import match_template as mt
from match_template import non_max_suppression
from match_template import matchTemplateMulti

import imagenes_util as img_util
import glob
```

### coca_multi.png

```python
img_name = 'coca_multi.png'
img = cv.imread('selected/' + img_name)
img_rgb= cv.cvtColor(img, cv.COLOR_BGR2RGB)

resultado, salida = matchTemplateMulti(img_rgb, metodo=cv.TM_CCORR_NORMED
                                       , umbral_coincidencia = 0.65
                                       , umbral_superposicion = .25
                                       , cual_imagen=mt.IMG_GRAY
                                       , cual_template=mt.TEMPLATE_INVERTIDO
                                              )
plt.figure(figsize=[10,8]); 
plt.imshow(salida); plt.title(f'{img_name} - Múltiples detecciones')
plt.show()

# Guardo el resultado
img_util.save_image('matched_multi/' + img_name, salida, img_rgb.shape)
```

```python
# Imágenes
image_names = glob.glob('matched_multi/*')
image_names.sort()
    
output = dict()

for image_name in image_names:
    img_name_short = image_name[8:]
    img = cv.imread(image_name,cv.IMREAD_COLOR)
    # Para plotear pasamos la imagen a RGB
    img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
    output[img_name_short] = img

img_util.plot_set_imagenes(output, '', columnas = 2)
```
