```python
import cv2 as cv
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

from match_template import matchTemplateMulti
import imagenes_util as img_util
```

### coca_multi.png

```python
img_name = 'coca_multi.png'

img = cv.imread('images/' + img_name)
img_rgb= cv.cvtColor(img, cv.COLOR_BGR2RGB)

resultado, salida, boxes = matchTemplateMulti(img_rgb, start=360)

if boxes>0:
    plt.figure(figsize=[10,8]); 
    plt.imshow(salida); plt.title(f'{img_name} - Múltiples detecciones ({boxes})')
    plt.show()
    
    # Guardo el resultado
    img_util.save_image('matched_multi/' + img_name, salida, img_rgb.shape)
else:
    print('No se detectaron logos en la imagen :(')
```
