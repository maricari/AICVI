```python
import cv2 as cv
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (9.0, 9.0)
```

## Histogramas

```python
img1 = cv.imread('images/img1_tp.png',cv.IMREAD_GRAYSCALE)
img2 = cv.imread('images/img2_tp.png',cv.IMREAD_GRAYSCALE)
plt.figure(figsize=[10,4])
plt.subplot(121), plt.imshow(img1, cmap='gray'); plt.title('Imagen 1')
plt.subplot(122), plt.imshow(img2, cmap='gray'); plt.title('Imagen 2')
plt.show()


# Propuesta de número de bins a usar
bins = [256,125,50,25,10,3]

n = len(bins)
i = 1
plt.figure(figsize=[10, n * 4])

for idx, b in enumerate(bins):
    hist, bins = np.histogram(img1.ravel(),b)
    plt.subplot(n,2,i); plt.hist(bins[:-1], bins, weights=hist)
    plt.title(f'{b} Bins')

    hist, bins = np.histogram(img2.ravel(),b)
    plt.subplot(n,2,i+1); plt.hist(bins[:-1], bins, weights=hist)
    plt.title(f'{b} Bins')

    i=i+2

plt.show()
    
```

### Conclusiones

- Los histogramas para ambas imágenes se ven exactamente iguales, no importa cuántos bins usemos.
- De los diferentes histogramas propuestos, el de 50 bins es el que mejor muestra cómo se distribuyen los distintos grises.
    - En el caso de la primera imagen, se ve la separación en tres grupos de gradientes: grises oscuros/negro, grises medios, y blanco.
    - En el caso de la segunda imagen, esa misma separación permite distinguir el fondo de la imagen (grises oscuros), el centro de la margarita (grises medios) y los pétalos (blanco).
- En conclusión, en este caso los histogramas son una herramienta útil para considerar como feature y entrenar un modelo de clasificación/detección de imágenes.
